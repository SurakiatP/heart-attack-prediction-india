# Exploratory Data Analysis
This notebook is used for initial exploration of the Heart Attack Prediction dataset.


```python
import pandas as pd

# Load the data
data_path = '../data/raw/heart_attack_prediction_india.csv'
df = pd.read_csv(data_path)
print(df.head())
print(df["Hypertension"].value_counts())
```

```python
# Display statistical summary
df_non_patient_id = df.drop(columns="Patient_ID")
df_non_patient_id.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(8, 5))

# Histogram for Age distribution
sns.histplot(df['Age'], bins=20, kde=True, color="blue")

# Labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(6, 4))

# Count the number of Male and Female
sns.countplot(data=df, x="Gender", palette="pastel")

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 5))

# Histogram with Kernel Density Estimate (KDE)
sns.histplot(df['Cholesterol_Level'], bins=30, kde=True, color="red")

plt.title("Cholesterol Level Distribution")
plt.xlabel("Cholesterol Level")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 5))

# Scatter plot for Age vs Cholesterol Level
sns.scatterplot(data=df, x="Age", y="Cholesterol_Level", hue="Heart_Attack_Risk", palette="coolwarm")

plt.title("Age vs Cholesterol Level")
plt.xlabel("Age")
plt.ylabel("Cholesterol Level")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 5))

# Boxplot for Systolic Blood Pressure and Diastolic Blood Pressure
sns.boxplot(data=df[['Systolic_BP', 'Diastolic_BP']], palette="pastel")

plt.title("Blood Pressure Distribution")
plt.ylabel("Blood Pressure")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
numeric_df = df_non_patient_id.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6, 4))

# Count plot for Heart Attack Risk
sns.countplot(data=df, x="Heart_Attack_Risk", palette="Set2")

plt.title("Heart Attack Risk Distribution")
plt.xlabel("Heart Attack Risk (0 = No, 1 = Yes)")
plt.ylabel("Count")

# Show the plot
plt.show()
```
